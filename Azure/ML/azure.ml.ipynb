# Linear Regression | Azure | Machine Learning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

#Metrics for regression models
from sklearn.metrics import r2_score

#Saving models
from joblib import dump

#Load model selection
from sklearn.model_selection import train_test_split

#Load Regression ML model
from sklearn.linear_model import LinearRegression

#Load Azure Model and Workspace
from azureml.core import Workspace
ws = Workspace.from_config(path="./config/config.json")
from azureml.core.model import Model

#Load Customer data
customers = pd.read_csv("./data/customer.csv")

#Register depedencies on Azure
# from azureml.core.conda_dependencies import CondaDependencies

# dependencies = CondaDependencies()

# dependencies.set_python_version("3.6.6")
# dependencies.add_pip_package("scikit-learn==0.20.3")
# dependencies.add_pip_package("numpy==1.16.2")
# dependencies.add_pip_package("pandas==0.23.4")
# dependencies.add_pip_package("matplotlib==3.0.0")
# compute_config.environment.python.conda_dependencies = dependencies
```

## Choose features / Labels

```python
y = customers["Yearly Amount Spent"]

X = customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

## Split the Data

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Train the model

```python
model = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1)

model.fit(X_train, y_train)
```

## Save the model

```python
os.makedirs("outputs", exist_ok=True)
dump(value=model, filename="outputs/yamtsx.pkl")
```

## Register the model into the `Azure` Workspace

```python
azure_saved_model = Model.register(workspace=ws, model_path="./outputs/yamtsx.pkl", model_name="customer_spent_lr")
```
